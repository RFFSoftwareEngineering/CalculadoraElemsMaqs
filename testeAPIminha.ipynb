```python
import pandas as pd

urlteste = "/content/2021-04-14.txt"

teste = pd.read_fwf(urlteste)
display(teste)
```

```python
teste.replace(',','', regex = True, inplace = True)
teste.replace("Z", "", regex = True, inplace = True)
teste.replace("T", " ", regex = True, inplace = True)
teste.replace("0.0offOFF0", "OFF ", regex = True, inplace = True)
teste.dropna(how = "all", inplace = True)
teste.columns = ["Data e Hora", "Temperatura", "Umidade", "PressãoAtm", "Altitude", "VentoRPM", "DirVento"]
display(teste)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (18, 10))
sns.set()

hist1 = sns.histplot(x='Temperatura', data=teste, binwidth = 0.02, kde = True)
hist1.set_title("Histograma teste", fontsize = 22)
hist1.set_xlabel("teste", fontsize = 22)
hist1.set_ylabel("teste", fontsize = 22)

plt.show()
```
